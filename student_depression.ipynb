```python
import scipy.stats as stats

# Split data by gender
male = data[data['Gender'] == 'Male']['Depression']
female = data[data['Gender'] == 'Female']['Depression']

# Perform independent t-test
t_stat, p_value = stats.ttest_ind(male, female)
print(f"t-statistic: {t_stat}, p-value: {p_value}")

if p_value < 0.05:
    print("Reject the null hypothesis: Significant difference between genders.")
else:
    print("Fail to reject the null hypothesis: No significant difference.")
```

```python
# Example: Compare Depression levels across different Academic Pressure groups
high_pressure = data[data['Academic Pressure'] == 'High']['Depression']
low_pressure = data[data['Academic Pressure'] == 'Low']['Depression']

# Perform an independent t-test
t_stat, p_value = stats.ttest_ind(high_pressure, low_pressure)

print(f"T-statistic: {t_stat}, P-value: {p_value}")
```
