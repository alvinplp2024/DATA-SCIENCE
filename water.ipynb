```python
data=data.rename(columns={'Per Capita Water Use (Liters per Day)':'Per Capital Water Use (Liters per Day)'})
```

```python
data
```

```python
sns.lineplot(x='Year',
             y='Total Water Consumption (Billion Cubic Meters)',
             data=data)
plt.title('Water Consumption Trends Over Time')
plt.show()
```

```python
sns.pairplot(data, vars=['Per Capital Water Use (Liters per Day)', 
                       'Rainfall Impact (Annual Precipitation in mm)', 
                       'Groundwater Depletion Rate (%)'])
plt.show()
```

```python
data.columns
```

```python
data_corr=data[['Agricultural Water Use (%)', 'Industrial Water Use (%)',
       'Household Water Use (%)']].corr()

sns.heatmap(data_corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.boxplot(x='Country', 
            y='Per Capital Water Use (Liters per Day)', 
            data=data)
plt.title('Per Capital Water Use by Country')
plt.show()
```

```python
sns.scatterplot(x='Rainfall Impact (Annual Precipitation in mm)', 
                y='Groundwater Depletion Rate (%)', 
                #hue='Country', 
                data=data)
plt.title('Rainfall vs. Groundwater Depletion')
plt.show()
```
