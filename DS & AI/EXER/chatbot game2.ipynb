```python
def chatbot_game2():
    print("Welcome to the chatbot powered by ALVIN ONDIEKI")
    print("---------------------------------------------------------")
    print("Answer the questions to Find out about Spooky Mansion")
    print("---------------------------------------------------------")

    state = "start"

    while True:
        if state == "start":
            print("\nYou are in an entrance of a spooky mansion.")
            print("---------------------------------------------------------")
            print("Do you choose to \n'1=enter' the mansion, \n'2=explore' the garden that surrounds it, or \n'3=walk away'?")
            choice = input("> ").strip().lower()

            if choice == "1":
                state = "mansion"
            elif choice == "2":
                state = "garden"
            elif choice == "3":
                state = "walk_away"
            else:
                print("Invalid choice. Please choose 'enter', 'explore', or 'walk away'.")
                print("---------------------------------------------------------")
        
        
        elif state == "mansion":
            print("\nYou must havee courage go through the entrance...explore the mansion.")
            print("---------------------------------------------------------")
            print("The heavy door behind you, is shut, dont worry, move next")
            print("---------------------------------------------------------")
            print("  1. A winding spiral staircase leads upward ('stairs=1').")
            print("  2. A creaking door on your left, hinting at secrets ('door=2').")
            print("  3. A dark, descending passage takes you into a shadowy basement ('basement=3').")
            print("---------------------------------------------------------")
            print("Which path will you dare to choose?")
            choice = input("> ").strip().lower()

            if choice == "1":
                state = "stairs"
            elif choice == "2":
                state = "left_door"
            elif choice == "3":
                state = "basement"
            else:
                print("Invalid choice. Please choose '1=stairs', '2=door', or '3=basement'.")
                print("---------------------------------------------------------")

        # Ascending via the stairs: Two mysterious doors await.
        elif state == "stairs":
            print("\nGood choice lets see what up stairs........")
            print("---------------------------------------------------------")
            print("Here there are two doors, choose wisely")
            print("---------------------------------------------------------")
            print(" The door on the 'right'(right ==1 ) ")
            print(" The door on the 'left' (left == 2) ")
            print("---------------------------------------------------------")
            print("Which door do you choose to open?")
            choice = input("> ").strip().lower()
            
            if choice == "1":
                state = "treasure_room"
            elif choice == "2":
                state = "ghost"
            else:
                print("Invalid choice. Please type '1=right' or '2=left'.")
                print("---------------------------------------------------------")
                
        # The treasure room: Victory is achieved.
        elif state == "treasure_room":
            print("\nYou push open the door on the right...")
            print("---------------------------------------------------------")
            print("Congratulations, you have discovered the treasure room and won the adventure!")
            break
        
        # The ghost: A sudden, terrifying encounter ends your quest.
        elif state == "ghost":
            print("\nYou push open the door on the left...")
            print("---------------------------------------------------------")
            print("\nGhost Scares you away,....sorry but keep safe...")
            print("Game over.")
            break

        # The trap door: Choosing the left door in the mansion leads to an unexpected fall.
        elif state == "left_door":
            print("\nYou decideD to open the door to your left.")
            print("---------------------------------------------------------")
            print("Sorry, you fell into the trap, Sorry one again.")
            print("---------------------------------------------------------")
            print("It’s the end of your journey – Game over.")
            break
        
        # The basement: In the depths of darkness, a secret mechanism brings hope.
        elif state == "basement":
            print("\nDetermined, you are going to a dark basement.")
            print("---------------------------------------------------------")
            print("Hey, go slowly, dont stamble and pain yourself, youll find a way out.")
            print("---------------------------------------------------------")
            print("Congratulations, you have successfully escaped – You win!")
            break
        
        # The garden: Nature and mystery intertwine outside the mansion.
        elif state == "garden":
            print("\nYou escape going to the room and  you've decided to explore the garden.")
            print("---------------------------------------------------------")
            print("Here are two features that you can choose one:")
            print("  - A strange, glowing object resting atop a moss-covered stone pedestal ('1=object').")
            print("  - A mysterious ancient well, its depths obscured by creeping ivy ('2=well').")
            print("---------------------------------------------------------")
            print("Which will you investigate?")
            choice = input("> ").strip().lower()
            
            if choice == "1":
                state = "amulet"
            elif choice == "2":
                state = "fall_well"
            else:
                print("Invalid choice. Please choose '1=object' or '2=well'.")
        
        # The magical amulet: A blessing in shining form.
        elif state == "amulet":
            print("\nYou carefully approach the object")
            print("---------------------------------------------------------")
            print("You exit the garden")
            print("Congratulations, you win!")
            break
        
        # The well: Temptation leads to a disastrous fall.
        elif state == "fall_well":
            print("\nYou step closer to peer into its murky depths.")
            print("---------------------------------------------------------")
            print("You fell into into the well and drown")
            print("---------------------------------------------------------")
            print("You loose, game over!")
            break
        
        # Walking away: Sometimes caution wins, but leaves a lingering mystery.
        elif state == "walk_away":
            print("\nYou decide to walk away from the spooky mansion.")
            print("---------------------------------------------------------")
            print("You safe leave but leaves many good thing behind unexplored.")
            print("---------------------------------------------------------")
            print("Game over.")
            break
        
    print("\nThanks for playing the Spooky Mansion Adventure!")
```

```python
# Run the game
chatbot_game2()
```
