**DBSCAN clustering algorithm**

```python
from numpy import unique
from numpy import where
from matplotlib import pyplot
from sklearn.datasets import make_classification
from sklearn.cluster import DBSCAN

# Initialize the dataset
training_data, _ = make_classification(
    n_samples=1000,
    n_features=2,
    n_informative=2,
    n_redundant=0,
    n_clusters_per_class=1,
    random_state=4
)

# Define the DBSCAN model
dbscan_model = DBSCAN(eps=0.25, min_samples=9)

# Fit the model and get cluster labels in one step
dbscan_result = dbscan_model.fit_predict(training_data)

# Get all unique clusters (including noise, labeled as -1)
dbscan_clusters = unique(dbscan_result)

# Plot the DBSCAN clusters
for cluster in dbscan_clusters:
    # Get indices of points in this cluster
    index = where(dbscan_result == cluster)[0]  # [0] because where() returns a tuple
    
    # Make the plot
    pyplot.scatter(training_data[index, 0], training_data[index, 1], label=f'Cluster {cluster}')

# Show the plot with a legend
pyplot.legend()
pyplot.title("DBSCAN Clustering")
pyplot.show()
```

**Gaussian Mixture Model algorithm**

```python
from numpy import unique
from numpy import where
from matplotlib import pyplot
from sklearn.datasets import make_classification
from sklearn.mixture import GaussianMixture

# Initialize the dataset
training_data, _ = make_classification(
    n_samples=1000,
    n_features=2,
    n_informative=2,
    n_redundant=0,
    n_clusters_per_class=1,
    random_state=4
)

# Define the Gaussian Mixture model
gaussian_model = GaussianMixture(n_components=2)

# Fit the model
gaussian_model.fit(training_data)

# Predict cluster assignments
gaussian_result = gaussian_model.predict(training_data)

# Get all unique clusters
gaussian_clusters = unique(gaussian_result)

# Plot the clusters
for cluster in gaussian_clusters:
    # Get indices of points in this cluster
    index = where(gaussian_result == cluster)[0]  # [0] because where() returns a tuple
    
    # Plot the points
    pyplot.scatter(training_data[index, 0], training_data[index, 1], label=f'Cluster {cluster}')

# Add legend and title
pyplot.legend()
pyplot.title("Gaussian Mixture Clustering")
pyplot.show()
```

**Affinity Propagation clustering algorithm**

```python
from numpy import unique
from numpy import where
from matplotlib import pyplot
from sklearn.datasets import make_classification
from sklearn.cluster import AffinityPropagation

# initialize the data set we'll work with
training_data, _ = make_classification(
    n_samples=1000,
    n_features=2,
    n_informative=2,
    n_redundant=0,
    n_clusters_per_class=1,
    random_state=4
)

# define the model
model = AffinityPropagation(damping=0.7)

# train the model
model.fit(training_data)

# assign each data point to a cluster
result = model.predict(training_data)

# get all of the unique clusters
clusters = unique(result)

# plot the clusters
for cluster in clusters:
    # get data points that fall in this cluster
    index = where(result == cluster)
    # make the plot
    pyplot.scatter(training_data[index, 0], training_data[index, 1])

# show the plot
pyplot.show()
```

**OPTICS algorithm**

```python
from numpy import unique
from numpy import where
from matplotlib import pyplot
from sklearn.datasets import make_classification
from sklearn.cluster import OPTICS

# Initialize the dataset
training_data, _ = make_classification(
    n_samples=1000,
    n_features=2,
    n_informative=2,
    n_redundant=0,
    n_clusters_per_class=1,
    random_state=4
)

# Define the OPTICS model
optics_model = OPTICS(min_samples=10, xi=0.05, min_cluster_size=0.05)

# Fit and predict cluster labels
optics_result = optics_model.fit_predict(training_data)

# Get all unique clusters (including noise points labeled as -1)
optics_clusters = unique(optics_result)  # Corrected: using optics_result

# Plot the clusters
for cluster in optics_clusters:
    # Get indices of points in this cluster
    index = where(optics_result == cluster)[0]  # [0] because where() returns a tuple
    
    # Plot the points
    pyplot.scatter(training_data[index, 0], training_data[index, 1], 
                   label=f'Cluster {cluster}' if cluster != -1 else 'Noise')

# Add legend and title
pyplot.legend()
pyplot.title("OPTICS Clustering")
pyplot.show()
```
