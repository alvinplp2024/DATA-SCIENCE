```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Training-ready_Production-ready-features-engineered.csv')
df
```

```python
# DATETIME INDEX
df['date_time'] = pd.to_datetime(df['date_time'])
df = df.set_index('date_time').sort_index()
df
```

```python
df.columns
```

```python
# 1. target
y = df['kwh']

# 2. features: drop multiple columns in one go
cols_to_drop = ['kwh', 'season', 'kwh_sqrt', 'kwh_yeo']
X = df.drop(columns=cols_to_drop)
```

```python
cols_to_prune = [
    'Snow_Cover_(%)_lag_24', 'Snow_Cover_(%)_lag_168',
    'Snow_Cover_(%)_roll_mean_3','Snow_Cover_(%)_roll_std_3',
    'Snow_Cover_(%)_roll_mean_24','Snow_Cover_(%)_roll_std_24',
    'Snow_Cover_(%)_roll_mean_168','Snow_Cover_(%)_roll_std_168',
    'Snow_Cover_(%)_yeo','current_yeo'
]

X_pruned = X.drop(columns=cols_to_prune)
```

```python
X_pruned.columns
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
# 1. pick only the numeric columns
numeric_cols = X_pruned.select_dtypes(include='number').columns
X_num = X_pruned[numeric_cols]

# 2. fit & transform
scaler = StandardScaler()
X_num_scaled = scaler.fit_transform(X_num)

# 3. if you need a DataFrame again:
X_scaled = pd.DataFrame(
    X_num_scaled,
    index=X_pruned.index,
    columns=numeric_cols
)
```

```python
# Split data
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

# **XGBOOST Model**

```python
# Initialize models
xgb_model = XGBRegressor(random_state=42)

# Fit models
xgb_model.fit(X_train, y_train)

# Predictions
xgb_preds = xgb_model.predict(X_test)
```

```python
# Evaluation function
def evaluate_model(y_true, y_pred, model_name):
    mae = mean_absolute_error(y_true, y_pred)
    # Calculate RMSE using NumPy for compatibility
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    r2 = r2_score(y_true, y_pred)
    print(f"{model_name} Evaluation:")
    print(f"  MAE : {mae:.4f}")
    print(f"  RMSE: {rmse:.4f}")
    print(f"  R²  : {r2:.4f}")
    print("-" * 30)

# Evaluate XGBoost models
evaluate_model(y_test, xgb_preds, "XGBoost")
```

## **XGBoost hyperparameter tuning**

```python
from sklearn.model_selection import GridSearchCV
from xgboost import XGBRegressor

# Define hyperparameter grid for XGBoost
param_grid_xgb = {
    'n_estimators': [100, 200],
    'max_depth': [3, 6, 10],
    'learning_rate': [0.01, 0.1, 0.2],
    'subsample': [0.8, 1.0]
}

# Initialize model
xgb = XGBRegressor(random_state=42)

# Grid search
grid_xgb = GridSearchCV(xgb, param_grid_xgb, cv=5, scoring='neg_mean_absolute_error', n_jobs=-1)
grid_xgb.fit(X_train, y_train)

# Best estimator and parameters
best_xgb = grid_xgb.best_estimator_
print("Best XGBoost Parameters:", grid_xgb.best_params_)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot feature importances
plt.figure(figsize=(24, 12))
importance_values = best_xgb.feature_importances_
sns.barplot(x=importance_values, y=numeric_cols)
plt.title("XGBoost Feature Importances")
plt.xlabel("Importance Score")
plt.ylabel("Feature")
plt.tight_layout()
plt.savefig('XGBoost_FeatureImportance_After_Pruning.png')
plt.show()
```

## **Visualize The predictions**

```python
import matplotlib.pyplot as plt

# 1. Generate predictions
y_pred = best_xgb.predict(X_test)
```

```python
import os
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure the directory exists
plot_dir = "plots"
os.makedirs(plot_dir, exist_ok=True)

def plot_model_performance(y_true, y_pred, model_name):
    # If passed a model object, grab its class name; otherwise assume it's already a string
    label = model_name.__class__.__name__ if not isinstance(model_name, str) else model_name

    # === Plot 1: Actual vs. Predicted ===
    plt.figure(figsize=(8, 6))
    residuals = y_true - y_pred

    sns.scatterplot(
        x=y_true,
        y=y_pred,
        hue=residuals,
        palette="coolwarm",
        size=residuals.abs(),
        sizes=(20, 100),
        alpha=0.7,
        edgecolor="black",
        linewidth=0.3,
        marker="o"
    )

    # 45° line
    mn, mx = min(y_true.min(), y_pred.min()), max(y_true.max(), y_pred.max())
    plt.plot([mn, mx], [mn, mx], "r--", lw=2, label="Ideal Fit")

    plt.xlabel("Actual Values (kWh)", fontsize=22, fontweight="bold")
    plt.ylabel("Predicted Values (kWh)", fontsize=22, fontweight="bold")
    plt.xticks(fontsize=16.5)
    plt.yticks(fontsize=16.5)
    plt.title(f"{label} — Actual vs. Predicted", fontsize=22, fontweight="bold")
    plt.legend(fontsize=13)
    plt.grid(True, linestyle="--", linewidth=0.7)
    plt.tight_layout()
    plt.savefig(os.path.join(plot_dir, f"{label}_actual_vs_predicted.tiff"), dpi=300)
    plt.close()

    # === Plot 2: Residual Distribution ===
    plt.figure(figsize=(8, 6))
    sns.histplot(residuals, bins=30, kde=True, color="skyblue")
    plt.axvline(0, color="red", linestyle="--", lw=2, label="Zero Residual")
    plt.xlabel("Residuals (Actual − Predicted)", fontsize=14, fontweight="bold")
    plt.ylabel("Count", fontsize=14, fontweight="bold")
    plt.title(f"{label} — Residuals Distribution", fontsize=18, fontweight="bold")
    plt.legend()
    plt.grid(True, linestyle="--", linewidth=0.5)
    plt.tight_layout()
    plt.savefig(os.path.join(plot_dir, f"{label}_residuals.png"), dpi=300)
    plt.close()

# Example usage:
plot_model_performance(y_test, y_pred, best_xgb)
# or
# plot_model_performance(y_test, y_pred, "XGBRegressor")
```
