# **1.3. Missing Value Treatment**

- **Interpolation for Continuous Variable**

**Why?**

- Interpolation estimates missing values based on previous and next values.  
- Works well for time-series data.


```python
df["v_red"] = df["v_red"].interpolate()
df["v_blue"] = df["v_blue"].interpolate()
df["v_yellow"] = df["v_yellow"].interpolate()
```

```python
df.isnull().sum()
```

```python
df["v_blue"]
```

## Extrapolation within bounds

- **Handles edge cases (start and end NaNs)**

```python
df["v_blue"] = df["v_blue"].interpolate(method="linear", limit_direction="both")
df["v_yellow"] = df["v_yellow"].interpolate(method="linear", limit_direction="both")
```

```python
df.isnull().sum()
```

# **1.4. Computing Hourly Averages**

- ## Dataset 1 (5min ---> 1h---> Frequency)

- There are 12 times 5min in an hour (1h = 5 times 12 = 60min)

- **Averages must be computed every 12 steps**

```python
# Set the datetime column as index
df.set_index('date_time', inplace=True)

# Resample the data to hourly frequency and compute the mean
df_hourly = df.resample('1h').mean()

# Reset index
df_hourly = df_hourly.reset_index()
df_hourly
```
