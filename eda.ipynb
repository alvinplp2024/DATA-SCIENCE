```python
data = df[['coaster_name',
    #'Length', 'Speed',
    'Location', 'Status',
    #'Opening date',
    #   'Type',
    'Manufacturer',
    #'Height restriction', 'Model', 'Height',
    #  'Inversions', 'Lift/launch system', 'Cost', 'Trains', 'Park section',
    # 'Duration', 'Capacity', 'G-force', 'Designer', 'Max vertical angle',
    #   'Drop', 'Soft opening date', 'Fast Lane available', 'Replaced',
    #  'Track layout', 'Fastrack available', 'Soft opening date.1',
    # 'Closing date',
    #'Opened',
    #'Replaced by', 'Website',
    #  'Flash Pass Available', 'Must transfer from wheelchair', 'Theme',
    #   'Single rider line available', 'Restraint Style',
    #   'Flash Pass available', 'Acceleration', 'Restraints', 'Name',
       'year_introduced', 'latitude', 'longitude', 'Type_Main',
       'opening_date_clean',
    #'speed1', 'speed2', 'speed1_value', 'speed1_unit',
       'speed_mph',
    #'height_value', 'height_unit',
    'height_ft',
       'Inversions_clean', 'Gforce_clean']].copy()
```

```python
data.shape
```

```python
data['opening_date_clean'] = pd.to_datetime(data['opening_date_clean'], format='mixed')
```

```python
data.dtypes
```

```python
data.columns
```

```python
data = data.rename(columns={'coaster_name':'Coaster_Name',
                    'year_introduced':'Year_Introduced',
                    'opening_date_clean':'Opening_Date',
                    'speed_mph':'Speed_mp',
                    'height_ft':'Height_ft',
                    'Inversions_clean':'Inversions',
                    'Gforce_clean':'Gforce'})
```

```python
data.head()
```

```python
data = data.drop_duplicates()
```

```python
data.duplicated().sum()
```

```python
data=data.loc[~data.duplicated(subset=['Coaster_Name','Location','Opening_Date'])] \
    .reset_index(drop=True).copy()
```

```python
data['Year_Introduced'].value_counts() \
    .head(10) \
    .plot(kind='bar', title='Test', color='b')
plt.ylabel('counts')
```

```python
data['Speed_mp'].plot(kind='hist',
                           title='Frequency coasters in mph',
                           color='green',
                           bins=20)
plt.xlabel('Frequency (mph)')
plt.show()
```

```python
data['Speed_mp'].plot(kind='kde',
                           title='Frequency coasters in mph',
                           color='green')
plt.xlabel('Frequency (mph)')
plt.show()
```

```python
data.plot(kind='scatter',
          x='Speed_mp',
          y='Height_ft',
          color='g',
          title='Speed vs Height')
plt.show()
```

```python
sns.scatterplot(x='Speed_mp',
                y='Height_ft',
                color='green',
                hue='Year_Introduced',
                data=data)
```

```python
sns.barplot(x='Speed_mp',
            y='Height_ft',
            color='green',
            data=data)
```

```python
sns.pairplot(data, vars=['Year_Introduced', 'Speed_mp',
                        'Height_ft','Inversions', 'Gforce'],
            hue='Type_Main')
plt.show()
```
